```python
import matplotlib.pyplot as plt
import random

class Location:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.location_name = "Default"
        self.distance = 0       # Used for Exercise 3, Task 3. Set to zero by default. 
        
    
    def get_coordinates(self):
        return f"Coordinates (x, y) are: ({self.x}, {self.y})."
        
    def set_coordinates(self, x, y):
        self.x = x
        self.y = y
        self.location_name = '' # This variable is included for testing purposesin order to assign a letter 
                                # to the location as per the sample output for Exercise 1, Task 3.
        
    def __str__(self):
        #return f"\nX: {self.x}, Y: {self.y}, Location Name: {self.location_name}" Line used for testing only.
        return f"\nX: {self.x}, Y: {self.y}"
# TASK 2: Implement a method called manhattan_dist that calculates the Manhattan distance between
#.        two locations.

    def manhattan_distance(self, other):
        self.distance = abs(self.x - other.x) + abs(self.y - other.y)
        return f"\nDistance of Reference Location from {other.location_name}: {self.distance}"
    
    def set_location_name(self, letter):
        self.location_name = letter
```

```python
# TASK 1: Implement the class constructor, and the suitable magic method that will be called when the
#.        restaurant object is printed.

class Restaurant:
    
    def __init__(self, name, location, opening_time, closing_time):
        self.name = name
        self.location = Location(location[0], location[1])
        self.opening_time = opening_time
        self.closing_time = closing_time
        
    def __str__(self):
        return f"\nRestaurant: {self.name}, Location (X,Y): ({self.location.x}, {self.location.y}), Opening Hours: {self.opening_time}:00 - {self.closing_time}:00"

# TASK 2: Implement the following additional methods. 

    def get_name(self):
        return f"Restaurant Name: {self.name}"
    
    def get_opening_hours(self):
        opening_hours = (self.opening_time, self.closing_time)
        return opening_hours
    
    def change_opening_hours(self, opening_time, closing_time):
        if opening_time < closing_time:
            self.opening_time = opening_time
            self.closing_time = closing_time
        
        else:
            print('\nError. Opening time must be earlier than closing time!\n')
            
    def getLocation(self):
        return self.location
    
    def is_open(self, current_time):
        if (current_time >= self.opening_time) and (current_time < self.closing_time):
            return True
        else: 
            return False
```

```python
class Where2Eat_tool:
     
    def __init__(self, x, y):
        self.restaurants = []
        self.ratings = dict()
        self.user_location = Location(x, y)
        self.open_restaurants = []
        self.open_restaurant_distances = []
        self.closed_restaurants = []
        
        # Lists for coordinates of open and closed restaurants used in Exercise 3, Task 4.
        self.x1 = []   # Open restaurants.
        self.y1 = []
        self.open_labels = []
        self.x2 = []   # Closed restaurants.
        self.y2 = []
        self.closed_labels = []
        self.x3 = []   # User location.
        self.y3 = []
        self.user_label = []
        
    def __str__(self):
        return f"Restaurant Object, {self.user_location}"
         
    # TASK 2 -  Create the method add_restaurant(new_rest, rating) that takes a restaurant object (as defined
    #           in Exercise 2) and a rating value and adds them into the internal data structures (i.e., to the list of restaurants
    #           and the dictionary of ratings ).
    #           Then create the method add_many_restaurants(rest_list,rating_list) that takes a list of restaurants and
    #           a list of ratings and adds them all to the internal data structures.
    #           Finally, create the method info that prints every stored restaurant together with its rating.
    
    def reload_location(self, x, y):
        self.user_location.set_coordinates(x, y)
        self.x3.clear()
        self.y3.clear()
        self.x3.append(x)
        self.y3.append(y)

    def add_restaurant(self, new_restaurant, rating):
        self.restaurants.append(new_restaurant)
        self.ratings[new_restaurant.name] = rating
       
    def add_many_restaurants(self, restaurant_list, rating_list):
        for restaurant in restaurant_list:
            self.restaurants.append(restaurant)
            self.ratings[restaurant.name] = rating_list[restaurant_list.index(restaurant)]
    
    def info(self):
        for restaurant in self.restaurants:
            print(restaurant)
            print('Rating: ', self.ratings[restaurant.name])

    def where_to_eat(self, current_time):
        
        self.open_restaurants.clear()               # Clear these lists at start of where_to_eat call in case of multiple calls and user location
        self.open_restaurant_distances.clear()      # changes between calls.
        self.closed_restaurants.clear()
        self.x1.clear() 
        self.y1.clear()
        self.open_labels.clear()
        self.x2.clear() 
        self.y2.clear()
        self.closed_labels.clear()
        self.x3.clear()
        self.y3.clear()
        self.user_label.clear()
        
        self.x3.append(self.user_location.x)
        self.y3.append(self.user_location.y)
        
        for restaurant in self.restaurants:
            if restaurant.is_open(current_time):
                self.open_restaurants.append(restaurant)    # Creating a list of open restaurants.
                self.x1.append(restaurant.location.x)
                self.y1.append(restaurant.location.y)
                self.open_labels.append(f"{restaurant.name}\nRating:{self.ratings[restaurant.name]}")

                
            else:
                self.closed_restaurants.append(restaurant) # Creating list of closed restaurants.
                self.x2.append(restaurant.location.x)
                self.y2.append(restaurant.location.y)
                self.closed_labels.append(f"{restaurant.name}\nRating:{self.ratings[restaurant.name]}")
        
        
        if len(self.open_restaurants) == 0:
             print('\nNo open restaurants.\n')
                
        else:
            
            for restaurant in self.open_restaurants:
                
                # Calculate distance of user location from each restaurant instance in open_restaurants.
                self.user_location.manhattan_distance(restaurant.location)      
                
                # Append this to list of restaurant distances. Index for a restaurant in open_restaurants should match
                # the index for it's distance in open_restaurant_distances.
                self.open_restaurant_distances.append(self.user_location.distance) 
                
            # Get minimum distance from open_restaurant_distances.
            min_distance = min(self.open_restaurant_distances)
            
            # Count to see if there is a tie for closest restaurant.
            min_distance_count = self.open_restaurant_distances.count(min_distance)
            #print('\nMinimum Distances Count:', min_distance_count, '\n')     # Test code: Only printed for observing how code works. 
            
            if min_distance_count == 1:
                selected_restaurant = self.open_restaurants[self.open_restaurant_distances.index(min_distance)]
                self.map()
                return (selected_restaurant, self.ratings[selected_restaurant.name], min_distance)
            
            elif min_distance_count > 1:
                
                # Extract restaurants causing the tie into another list. 
                tiebreaker_restaurant_list = []
                print(tiebreaker_restaurant_list)
                
                # Make list of restaurants tied for minimum distance from user. 
                for distance in self.open_restaurant_distances:
                    if distance == min_distance:
                        
                        #Using index information from distances list to get the restaurant object from open_restaurants. 
                        tiebreaker_restaurant_list.append(self.open_restaurants[self.open_restaurant_distances.index(distance)])
                print(tiebreaker_restaurant_list)
                
                # Make a list of rankings of restaurants tied for minimum distance.
                tiebreaker_rating_list = []
                
                # Using index of min distance value in distances list to get the restaurant object to in turn get it's name and use that to get the
                # rating from the dictionary.
                
                for distance in self.open_restaurant_distances:
                    if distance == min_distance:
                        tiebreaker_rating_list.append(self.ratings[self.open_restaurants[self.open_restaurant_distances.index(distance)].name])
                        
                print(tiebreaker_rating_list)   # Only printed for observing how code works. 
                
                rating_max = max(tiebreaker_rating_list)
                print(rating_max)               # Only printed for observing how code works. 
                
                selected_restaurant = tiebreaker_restaurant_list[tiebreaker_rating_list.index(rating_max)]
                self.map()
                return (selected_restaurant, self.ratings[selected_restaurant.name], min_distance)
        
        # Create scatterplot to display open, closed and selected restaurants.
        # Variables pairs (x1,y1), (x2, y2) and (x3,y3) represent the coordinates for open, closed and
        # selected restaurant respectively.
        
    def map(self):
        plt.title('Restaurant Selection')
        plt.scatter(self.x1, self.y1, label='Scatter Plot', color='green', marker='o')
        plt.scatter(self.x2, self.y2, label='Scatter Plot', color='red', marker='o')
        plt.scatter(self.x3, self.y3, label='Scatter Plot', color='yellow', marker='o', s = 200)
        plt.scatter(self.x3, self.y3, label='Scatter Plot', color='blue', marker='o')
        
        # Note: The following code was sourced from:
        # https://saturncloud.io/blog/scatter-plot-labels-in-one-line-matplotlib/
        
        for (i, j) in zip(self.x1, self.y1):
            plt.text(i, j, f"{self.open_labels[self.x1.index(i)]}", fontsize = "8")
            
        for (i, j) in zip(self.x2, self.y2):
            plt.text(i, j, f"{self.closed_labels[self.x2.index(i)]}", fontsize = "8")
            
        plt.text(self.x3[0], self.y3[0], "User Location", fontsize = "8")
        
        plt.xlabel('X-Coordinate')
        plt.ylabel('Y-Coordinate')
        plt.xlim(-100, 100)
        plt.ylim(-100, 100)
        plt.show()

        
            
            
```
